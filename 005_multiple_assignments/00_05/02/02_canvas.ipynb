```python
import tkinter as tk #use instead of canvas graphics

# Constants
CELL_SIZE = 30
CANVAS_WIDTH = 800
CANVAS_HEIGHT = 750
ERASER_SIZE = 30

def create_grid(canvas):
    """Create a grid of blue rectangles on the canvas."""
    rows = CANVAS_HEIGHT // CELL_SIZE
    cols = CANVAS_WIDTH // CELL_SIZE
    for row in range(rows):
        for col in range(cols):
            x1 = col * CELL_SIZE
            y1 = row * CELL_SIZE
            x2 = x1 + CELL_SIZE
            y2 = y1 + CELL_SIZE

            
            canvas.create_rectangle(x1, y1, x2, y2, fill="#0a29f5", outline="#fefbfa", tag="cell")

def erase_cells(canvas, eraser_coords):
    """Turn blue cells to white if they overlap with the eraser."""
    x1, y1, x2, y2 = eraser_coords
    overlapping = canvas.find_overlapping(x1, y1, x2, y2)
    for item in overlapping:
        if "cell" in canvas.gettags(item):
            canvas.itemconfig(item, fill="#fefbfa")

def main():
    root = tk.Tk()
    root.title("Eraser Canvas")

    canvas = tk.Canvas(root, width=CANVAS_WIDTH, height=CANVAS_HEIGHT)
    canvas.pack()

    create_grid(canvas)

    # pink eraser
    eraser = canvas.create_oval(0, 0, ERASER_SIZE, ERASER_SIZE, fill="#cc88e3", outline="#803799",width="2")

    def on_mouse_move(event):
        x = event.x
        y = event.y
        # Update eraser position
        canvas.coords(eraser, x, y, x + ERASER_SIZE, y + ERASER_SIZE)
        # Check for overlap and erase cells
        erase_cells(canvas, (x, y, x + ERASER_SIZE, y + ERASER_SIZE))

    # Bind the mouse movement to eraser behavior
    canvas.bind("<Motion>", on_mouse_move)

    root.mainloop()

if __name__ == '__main__':
    main()
```
