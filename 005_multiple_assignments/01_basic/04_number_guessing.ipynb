```python
import random
import time

random_number= random.randint(1,50)

win = True
guess_count = 0
print("-------------------------------------------------------------------------")
print('Number guessing Game,"Guess number from 1 to 50 and win prize "')
time.sleep(0.8)
while win == True:
      input_value = input("guess?:  ")


      if input_value.isdigit() == True:
            print(f"🔵hmm hmm so you guess {input_value}")
            convert = int(input_value)
            if convert <= 50:
                if  convert == random_number:
                  for i in range(2):
                          time.sleep(0.3)
                          print("\r⏳processing.",end="" )
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing.",end="")
                          time.sleep(0.2)
                  print(f"\n🟢You win! {random_number}=={convert}")
                  print(f"\n🎯 Total guesses made: {guess_count}")
                  print("-------------------------------------------------------------------------")
                  win =False
                  break

                elif convert > random_number:
                  for i in range(2):
                          time.sleep(0.3)
                          print("\r⏳processing.",end="" )
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing.",end="")
                          time.sleep(0.2)
                  print(f"\n🔴 Wrong: ??? ≠ {convert}")
                  print("🔼 Your guess was higher than the secret number!")
                  guess_count += 1
                  
                  print(f"\n🎯 Total guesses made: {guess_count}")
                  win = True

                else:
                  for i in range(2):
                          time.sleep(0.3)
                          print("\r⏳processing.",end="" )
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing...",end="")
                          time.sleep(0.3)
                          print("\r⏳processing..",end="")
                          time.sleep(0.3)
                          print("\r⏳processing.",end="")
                          time.sleep(0.2)
                  print(f"\n🔴 Wrong: ??? ≠ {convert}")
                  print("🔽 Your guess was lower than the secret number!")
                  guess_count += 1
                  
                  print(f"\n🎯 Total guesses made: {guess_count}")
                  win = True
            else:
              print(f"🚫{convert} is out of range")
      else:
        print(f"🚫{input_value} is not a number")



```
