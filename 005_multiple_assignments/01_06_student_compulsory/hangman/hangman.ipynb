```python
words = [
    "apple", "book", "chair", "table", "house", "plant", "phone", "shirt", "water", "bread",
    "pencil", "guitar", "jungle", "butter", "planet", "rocket", "summer", "window", "sister", "animal",
    "mystery", "kingdom", "picture", "beneath", "journey", "freedom", "library", "whisper", "diamond", "venture"
]
```

```python
import random

#i made on my own because tutorial was long and i am lazy
# enjoy game!
def choose_random_word():
    word = random.choice(words)
    
    return word


def acsii_man(x):

    guy = lives_visual_dict[x]
    return guy



def guess_word(gw):
    itrate = 0
    LIVES = 8
    guess= input("guess a word").lower()

    while guess != gw:

        itrate += 1
        life_left = 7-itrate 

        if itrate >= LIVES:
            print("you used all 7 lives!")
            print("Mission Failed")
            print(f"The answer was {gw}")
            break
            
            
        
        man = acsii_man(life_left)
        print("try again!")
        print(man)
        print(f"Lives life: {life_left} ")
        guess= input("guess a word").lower()

        

    if life_left != -1:
        print (f"you got it! answer is {gw}")

       

   


    

    

if  __name__ =="__main__":
    chosen = choose_random_word()
    guess_word(chosen)
```
